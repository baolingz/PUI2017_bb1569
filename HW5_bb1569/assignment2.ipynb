## AD and KS test

I need to figure out what each test function returns. Let me run them once testing an obvious result: is a Gaussian Gaussian? I know the answer so this is a reality check to make sure I understand the output of the functions (and read the manual at the same time!)

I am testing a normal distribution (Gaussian, mean = 0, std = 1) against a Gaussian model, and a Poisson with small $\lambda$, or a binomial with small np against a Gaussian model. I expect to find that in the first case the Gaussian is, well, Gaussian and in the second case that the binomial with small np is NOT Gaussian

```python
#generate the distribution
alpha = 0.05
np.random.seed(999)
dist_n = np.random.randn(1000) 

#test ad and ks. those are easy
print ("normal on normal", scipy.stats.kstest(dist_n,'norm'))
print ("normal on normal", scipy.stats.anderson(dist_n, dist='norm'))
print ("") 
```

###  ** H_0 ** : The tested distribution is not different than a Gaussian $ N(0,1) $ 

**KS**: high p-value pvalue>0.05:  H0 that sample comes from a Gaussian distribution cannot be rejected 

**AD**: at alpha = 0.05 the statistics = 0.26 is smaller than the 0.784 critical value for $\alpha$=0.05 (thresh_0.05 = 0.784): the H0 cannot be rejected.

##### (So the gaussian is indeed a gaussian ..)

```python
np.random.seed(999)
# poisson with 1000 elements, lambda=100 (high)
dist_p1 = np.random.poisson(100, 1000)
pl.figure()
pl.hist(dist_p1, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions in shape\nbut different mean and stretch")
pl.show()
print ("poisson on normal", scipy.stats.kstest(dist_p1,'norm'))
print ("poisson on normal", scipy.stats.anderson(dist_p1, dist='norm'))
print ("\n\nIn this case the means are different, " + 
       "so the distributions will be different for KS and AD test. " + 
       "I want a distribution with 'Poisson with high mean' shape cause that is more Gaussian " + 
       "but I want the mean to be the same as for my Gaussian: mean=0. "
       "Similarly, the standard deviation is sqrt(lambda) = sqrt(100). I want it to be 1") 


#So now we make a centered poisson, lambda=100

dist_p = (np.random.poisson(100, 1000) - 100) * 0.1 #-100 fixes mean, *0.1 fixes stdev
pl.figure()
pl.hist(dist_p, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions")
pl.show()

print ("poisson vs normal", scipy.stats.kstest(dist_p,'norm'))
print ("poisson vs normal", scipy.stats.anderson(dist_p, dist='norm'))
print ("KS: lower p-value but we still fail rejecting the H0 at alpha=0.05") 
print ("AD: statistics AD=0.92 larger than threshold (for alpha=0.05)") 

threshold = scipy.stats.anderson(dist_n, 
                                 dist='norm')[1][scipy.stats.anderson(dist_n, 
                                                                      dist='norm')[2]==[5.0]]
print ("thresh_0.05 = 0.784", threshold[0])

#binomial with n=10, p=0.5 => mean=np=5, stdev=sqrt(np(1-p))
n, p = 10, 0.5
dist_b = (np.random.binomial(n, p, 1000) - n*p) / np.sqrt(n*p*(1-p))  #subtract mean divide by stdev
pl.figure()
pl.hist(dist_b, alpha=0.5, label="Binomial")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions")
pl.show()

print ("binomial vs normal", scipy.stats.kstest(dist_b,'norm'))
print ("binomial vs normal", scipy.stats.anderson(dist_b, dist='norm'))
print (" what does the result of each test say??")
```

The statistics value will be **larger** if the distributions are **not** likely to be related. 
think about this in terms of rejection of the NULL hypothesis. 

*(BB_1569)*

* **Test1** $H_0=$ : A poisson with $\lambda$=100 is not different from a $ N(0,1) $ .   
    With a p-value close to 0, this null hypothesis is rejected at the $\alpha$=0.05 confidence level. I.e. the two distributions are statistically different. 
      
    
* **Test2** $H_0$  : A poisson with $\lambda$=100 fixed so that it has a mean of 0 and a std of 1 is not different from a $ N(0,1) $ .  
    We fail to reject the null hypothesis at the $\alpha$=0.05 confidence level (p-value=0.30 for K-S test, and for Anderson test we have a statistic of 0.92>0.784 thershold for $\alpha$=0.05. I.e. the two distributions are not statistically different. 
      
      
* **Test3** $H_0$ : A binomial distribution with n, p = 10, 0.5, standardize to mean=0 and std=1, is not different from a $N(0,1)$.
    With a p-value close to 0, we reject the null hypothesis at the $\alpha$=0.05 confidence level. I.e. the two distributions are statistically different.


## KL DIVERGENCE

The KL divergence is a little trickier: 

__ empirical distribution__ 
it wants to know the distribution _y_ value at a location _x_.  have to pass it the values of the _distribution_ at some x's, not my histogram though, si I need a  **NORMALIZED HOSTOGRAM OF MY DISTRIBUTION**, i.e. derive a probability distribution function (PDF) from the data. 

__model__
Then I have to pass it a function for my comparison distribution evaluated at the same _x_-values where my empirical distribution is evaluated (not just the name of a distribution like for KS and AD)

```python
# np.histogram returns the value of the histogram (y) and the edges of the bins (x). 
# Notice: the edges, not the centers
dist_p1 = (np.random.poisson(2, 1000) - 2)  # I tested also a poisson with lambda 2
#which would surely lead to rejection of null as to be sure my tests were working as expected.
distpdf_p1,mybins_p1, = np.histogram(dist_p1, density=True)
distpdf_n, mybins_n, = np.histogram(dist_n, density=True)
distpdf_b, mybins_b, = np.histogram(dist_b, density=True)
distpdf_p, mybins_p, = np.histogram(dist_p, density=True)

#notice the extra comma on the left side of the '=' sign: 
#that tells numpy take the first two values returned, 
#and throw away the rest
```

Then i have to pass it the functional form of a normal PDF, which is returned by the function scipy.stats.norm for example (a google search would have told you!) 
evaluated at the bin centers of the histogram from before

```python
#get the bin centers
bincenters_n = mybins_n[:-1] + 0.5*(mybins_n[1] - mybins_n[0])
bincenters_b = mybins_b[:-1] + 0.5*(mybins_b[1] - mybins_b[0])
bincenters_p = mybins_p[:-1] + 0.5*(mybins_p[1] - mybins_p[0])
bincenters_p1 = mybins_p1[:-1] + 0.5*(mybins_p1[1] - mybins_p1[0])

print ('the KL test returns the "entropy". ' + 
       "To figure out what the number means thoug in terms of NULL rejection, help yourself " + 
       "with the fact that you know what the result should be for these three distributions!")
print ("normal on normal", 
    scipy.stats.entropy(distpdf_n, 
                            scipy.stats.norm.pdf(bincenters_n)))
print ("poisson on normal", 
       scipy.stats.entropy(distpdf_p, 
                                scipy.stats.norm.pdf(bincenters_p)))

print ("binomial on normal", 
       scipy.stats.entropy(distpdf_b, 
                                scipy.stats.norm.pdf(bincenters_b)))

print("poisson lambda 2 on normal",
     scipy.stats.entropy(distpdf_p1, 
                                scipy.stats.norm.pdf(bincenters_p1)))

#you can interpret this as a distance: 
# it increases as the distributions diverge
```

Wikipedia page on KL test -- to understand the test better https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence. 

As the entropy value increases, the two compared distributions are more and more divergent from one another. Therefore, we see that the normal on normal, and poisson and normal give a very small KL divergence, suggesting the two distributions behave similarly. The KL divergence for the binomial on normal is 10 orders of magnitude larger, suggeting that there is some difference on how these two distributions behave. Similary, the Pois($\lambda=2 $) has a large KL value because it is very different from a standard normal.

# BINOMIAL - as N increases
with the binomial distribution I can change n*p. larger n*p will make the distribution look more "normal" (i.e. Gaussian). 
In a loop test that the binomial tends to a normal distribution as you increase the value of n*p.

```python
#initializing some arrays:
narray = np.arange(1,500,10)
ks_s_b = np.zeros(len(narray))
ks_p_b = np.zeros(len(narray))
ad_b = np.zeros(len(narray))
kl_b = np.zeros(len(narray))
chi_s_b = np.zeros(len(narray))
chi_p_b = np.zeros(len(narray))
```

```python
np.random.seed(910)

#loop over the different values of n (or n and p)
for i,n in enumerate(narray):
    p = 0.5 #parameter for the binomial, my arbitrary choice
    #generate the distribution
    dist = (np.random.binomial(n, p, 1000) - n * p ) / np.sqrt(n * p * (1-p))
    #run the tests. 
    ks_s_b[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_b[i] = scipy.stats.kstest(dist,'norm')[1]
    #print (scipy.stats.kstest(dist,'norm'))
    #print (scipy.stats.anderson(dist, dist='norm'))
    ad_b[i] = scipy.stats.anderson(dist, dist='norm')[0]
    #ad_b[i] = ad_b[i] / ad_b[0]
    #print ad_b[i]
  
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])


    #when i was coding this up something was wrong. i put some plots in to figure out what... 
    # just so you know.
    #if i%10 == 0: 
        #pl.hist(dist, bins=mybins)
        #pl.plot(bincenters, scipy.stats.norm.pdf(bincenters)*1000)
        
    kl_b [i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins,normed=True)[0], 
                                    scipy.stats.norm.pdf(bincenters))
    #print(np.histogram(dist,bins=mybins)[0]*1000)
    #print(scipy.stats.norm.pdf(bincenters)*1000)
    
    chi_s_b[i] = scipy.stats.chisquare(np.histogram(dist,bins=mybins,normed=True)[0]*100,
                                       scipy.stats.norm.pdf(bincenters)*100)[0]
    
    chi_p_b[i] = scipy.stats.chisquare(np.histogram(dist,bins=mybins,normed=True)[0]*100,
                                       scipy.stats.norm.pdf(bincenters)*100)[1]
    
    
```

# POISSON - as $\lambda$ increases

With the Poisson distribution I can change the parameter $\lambda$. A larger $\lambda$ will make the distribution look more "normal" (i.e. Gaussian). In a loop test that the Poisson tends to a normal distribution as you increase the value of $\lambda$

```python
#initialize arrays
larray = np.arange(1,500,10)
ks_s_p = np.zeros(len(larray))
ks_p_p = np.zeros(len(larray))
ad_p = np.zeros(len(larray))
kl_p=np.zeros(len(larray))
chi_s_b = np.zeros(len(larray))
chi_p_b = np.zeros(len(larray))
```

```python
#loop over different values of lambda
for i,lamb in enumerate(narray):
    #generate the distribution
    dist = (np.random.poisson(lamb, 1000) - lamb )*sqrt(1/lamb)
    #run the tests. 
    ks_s_p[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_p[i] = scipy.stats.kstest(dist,'norm')[1]
    #print (scipy.stats.kstest(dist,'norm'))
    #print (scipy.stats.anderson(dist, dist='norm'))
    ad_p[i] = scipy.stats.anderson(dist, dist='norm')[0]
    #ad_b[i] = ad_b[i] / ad_b[0]
    #print ad_b[i]
  
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])

    #when i was coding this up something was wrong. i put some plots in to figure out what... 
    # just so you know.
    #if i%10 == 0: 
    #    pl.hist(dist, bins=mybins)
    #pl.plot(bincenters, scipy.stats.norm.pdf(bincenters)*1000)
    kl_p [i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0], 
                                    scipy.stats.norm.pdf(bincenters) + lamb)
    chi_s_b[i] = scipy.stats.chisquare(np.histogram(dist,bins=mybins,normed=True)[0]*100,
                                       scipy.stats.norm.pdf(bincenters)*100)[0]
    
    chi_p_b[i] = scipy.stats.chisquare(np.histogram(dist,bins=mybins,normed=True)[0]*100,
                                       scipy.stats.norm.pdf(bincenters)*100)[1]
```

```python
fig = pl.figure(figsize = (12,16))
fig.add_subplot(221)
pl.plot(larray, ks_s_p, label='KS statistics')
pl.plot(larray, ks_p_p, label='KS p-value')
pl.legend()
pl.xlabel("$\lambda$",fontsize=14)
pl.title("KS test")

fig.add_subplot(222)
pl.plot(larray, ad_p,  label='AD statistics')
pl.plot([larray[0], narray[-1]],[threshold, threshold], label="treshold")
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.legend()
pl.xlabel("$\lambda$",fontsize=14)
pl.title("AD test")

fig.add_subplot(223)
pl.plot(larray, kl_p, label='K-L (entropy)')
pl.xlabel("$\lambda$",fontsize=14)
pl.legend()
pl.title("KL test");

fig.add_subplot(224)
pl.plot(larray, chi_s_b, label="Pearson Chi-Sq")
pl.plot(larray, chi_p_b, label="Peason p value")
pl.xlabel("$\lambda$",fontsize=14)
pl.title("Chi-Sq Test")
pl.legend()
pl.ylim(0,10);
```

**Figures 5, 6, 7, 8** These figures respectively represent the behaviour of the KS, AD, KL entropy, Pearson Chi-Sq entropy statistics, testing the null hypothesis that the Poisson Distribution is not different from the Gaussian, as $\lambda$ increases from 1 to 500. 

### Interpretation :  
In **Figure 5** , as $\lambda$ increases we see that the KS-statistic decreases, becoming lower than the threshold, and the p-value increases, going above the significance level of $\alpha=0.05$, or other selected confidence levels. As $\lambda$ increases, we are more likely to fail to reject the null hypothesis. Note that the p-value starts going above 0.05 sooner than for the binomial.
